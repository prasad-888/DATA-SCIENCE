### 2 Row 3 Column

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 11)
y= x**2


plt.subplot(2,3,6)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('SOme title')


plt.subplot(2,3,3)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('SOme title')

plt.subplot(2, 3, 1)
plt.plot(x,y,'r')
plt.title('red')
plt.show()
```

```python
fig = plt.figure()
# Add a set of axes to the figure
axes = fig.add_axes([0, 1, 0.8, 0.8])  # (left, bottom, width, height) ranges from 0 to 1
# Plot on the defined axes
axes.plot(x, y, 'b')  ## b here is blue
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set Y Label')
axes.set_title('Set Title')
plt.show()
```

### Adding multiple axes to a figure

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  ### MAin Axis
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])   ### Inset Axis


# Main plot
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Axes 1 Title')



# Inset plot
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title')
plt.show()
```

### Adjusting Layout with tight_layout()

A common challenge with subplots is overlapping  content, making it hard to read. The fig.tight_layout() method automatically adjusts the spacing between subplots

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

# Adjust layout

fig.tight_layout()
plt.show()
```

### Setting Plot ramges
The range of the plot can be adjusted using set_ylim() and set_xlim() or by using
axis('tight') for automatically fitted axes

```python
fig , axes = plt.subplots(1, 3, figsize=(12, 4))


# Default range
axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("Default Axes Range")


# Tight range
axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("Tight Axes")

# Custom rnage
axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("Custom Axes Range")

plt.show()

```
