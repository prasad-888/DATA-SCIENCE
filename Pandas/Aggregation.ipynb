## GROUPING MULTIPLE COLUMNS

```python
flights.groupby(['AIRLINE','WEEKDAY']) ['CANCELLED'].agg('sum')
```

```python
flights.groupby(['AIRLINE','WEEKDAY']) ['CANCELLED'].agg('sum').head(7)
```

```python
flights.groupby(['AIRLINE', 'WEEKDAY'])[['CANCELLED', 'DIVERTED']].agg(['sum', 'mean'])
```

```python
group_cols = ['ORG_AIR', 'DEST_AIR']
agg_dict = {
    'CANCELLED': ['mean', 'size'],  # Calculate mean and total number of flights
    'AIR_TIME': ['mean', 'var']     # Calculate average airtime and variance
}
flights.groupby(group_cols).agg(agg_dict).head()
```
