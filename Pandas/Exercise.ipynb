## Exercise 4: Create a new column TOTAL SALARY that is 10% higher than the current SALARY column, plus the BONUS amount. Use .loc on the left-hand side and ensure TOTAL SALARY is an integer data type.

```python
df['TOTAL SALARY']=df.loc[:,'SALARY']+df.loc[:,'SALARY']*0.1 + df.loc[:,'BONUS']

df['TOTAL SALARY'] = df['TOTAL SALARY'].astype(int)
df
```
