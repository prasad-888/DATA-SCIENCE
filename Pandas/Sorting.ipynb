## REPLACING VALUES

```python
import pandas as pd

df = pd.DataFrame({
    'A' : [0, 1, 2, 3, 4],
    'B': [5, 6, 7, 8, 9],
    'C': ['a','b','c','d','e'],
   
})
```

```python
df
```

```python
df.replace(0, 5)
```

```python
df.replace({0:10, 1:100 })
```

```python
## Specifying replacement by column
```

```python
df.replace({'A':0,'B':5},100)
```

```python
df.replace({'A':{0:100, 4: 400}})
```
