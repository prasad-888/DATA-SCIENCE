```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Categorical Data Plots
Seaborn provides various plots for visualizing categorical data, which can be particularly helpful for analyzing distributions, counts, and statistical summaries across categories. Here’s an overview of some of the main types of categorical plots:

barplot: Shows an aggregated value (like mean) for each category using rectangular bars.
countplot: Displays the count of each unique category as bars, making it ideal for frequency analysis.
boxplot: Summarizes the distribution of data across categories with boxes and whiskers to show spread, quartiles, and outliers.
violinplot: Combines aspects of boxplot and KDE plot, showing data distribution and density across categories.
stripplot: Plots individual observations for each category as dots, allowing for insight into data distribution.
swarmplot: Similar to a strip plot but adjusts the position of each dot to avoid overlap, providing a clearer view of the density and distribution of data points.
These plots offer a range of ways to analyze and visualize relationships and distributions within categorical data.

### BARPLOT

```python
tips = sns.load_dataset('tips')
tips.head()
```

## 1. Basic Barplot:
By default, sns.barplot() aggregates data based on the mean of the specified numeric feature for each category.

```python
sns.barplot(x='sex',y='total_bill', data=tips)
plt.show()
```

## 2. Using a Different Estimator:
You can specify a custom function as an estimator, allowing for different calculations, such as standard deviation.

```python
sns.barplot(x='sex',y='total_bill',data=tips , estimator=np.std)
plt.show()
```

### Countplot
A countplot in Seaborn is similar to a barplot but is specifically designed to count occurrences of each category within a single categorical variable. Since it’s a count, you only need to provide the x (or y) variable without an additional y variable.

```python
sns.countplot(x='sex',data=tips)
plt.show()
```

### Boxplot
Boxplots, also known as box-and-whisker plots, are useful for visualizing the distribution of data and identifying potential outliers. In a boxplot, the "box" represents the interquartile range (IQR) of the data, while the "whiskers" show the extent of the rest of the distribution, excluding points classified as outliers.

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips, palette='rainbow')
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='size',palette='rainbow')
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips, hue='sex', palette='rainbow')
plt.show()
```

```python
sns.boxplot(data=tips, palette='rainbow', orient='h')
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',hue='smoker',data=tips,palette='coolwarm')
plt.show()
```

### Violinplot
A violin plot combines elements of a box plot with a kernel density plot. It displays the distribution of data across various categories, similar to a box plot, but with the added benefit of a density estimation on each side, which helps visualize the data’s distribution shape.

```python
sns.violinplot(x='day',y='total_bill',data=tips,palette='rainbow')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',hue='sex',data=tips, palette='rainbow')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',hue='sex',data=tips,palette='Set1')
plt.show()
```

### STRIP PLOT

```python
sns.stripplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True) ## Jitter is for making sure no points  overlap
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True, hue='sex',palette='Set1')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True, hue='sex',palette='Set1', dodge=True)
plt.show()
```

### SWARM PLOTS

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex',palette='Set1',dodge=True)
plt.show()
```

### Combining Categorical Plots
Seaborn allows you to combine multiple categorical plots to better understand the data. For instance, you can layer a violinplot and a swarmplot to show both the distribution and individual data points.

```python
sns.violinplot(x="tip", y="day", data=tips, palette='rainbow')
sns.swarmplot(x="tip", y="day", data=tips, color='black', size=3)
plt.show()
```

### Catplot
catplot in Seaborn is a versatile function for creating categorical plots. By specifying the kind parameter, you can create a variety of plot types, such as bar, strip, swarm, box, and violin plots, all using a consistent syntax.

```python
sns.catplot(x='sex',y='total_bill',data=tips, kind='bar')
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=tips)
plt.show()
```
