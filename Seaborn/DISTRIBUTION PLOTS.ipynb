### DISTPLOT

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

tips = sns.load_dataset('tips')
tips
```

```python
sns.displot(tips)
plt.show()
```

```python
sns.displot(tips['total_bill'])
plt.show()
```

### Jointplot
The Jointplot allows you to match up two distplots for bivariate data, with various comparison parameters such as scatter, reg, resid, kde, hex, etc.

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='scatter')
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='resid')
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
plt.show()
```


### PAIR PLOT

The Pairplot plots pairwise relationships across an entire dataframe for numerical columns. It also supports a hue argument for color-coding based on categorical columns.

```python
sns.pairplot(iris)
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.pairplot(tips, hue='time')
plt.show()
```

```python
sns.pairplot(tips, hue='time',palette='coolwarm')
plt.show()
```

### Kdeplot
Kdeplots (Kernel Density Estimation plots) visualize the probability density of a continuous variable. In a KDE plot, each observation is represented by a Gaussian (normal) distribution centered on that observation's value, and the cumulative plot reflects the combined density across all observations.

```python
sns.kdeplot(tips['total_bill'])
plt.show()
```
