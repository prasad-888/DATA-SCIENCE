```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
tips[['total_bill','tip','size']].corr()
sns.heatmap(tips[['total_bill','tip','size']].corr())
plt.show()
```

###Heatmap
The heatmap function in Seaborn is useful for visualizing matrices of numerical values, such as correlation matrices or pivot tables. Let's explore how to create and customize heatmaps using Seaborn.

```python
sns.heatmap(tips[['total_bill','tip','size']].corr(), cmap='coolwarm',annot=True)
plt.show()
```
